```python
def sigmoid(z):
    return 1.0/(1.0+np.exp(-z))
```

```python
z = np.array([0.5,0.5])
sigmoid(z)
```

```python
def model_output(x,y,w,b):
    m = len(y)
    f_wb = np.zeros(m)
    f_wb = np.dot(x,w) + b
    z = sigmoid(f_wb)
    return z
```

```python
def compute_cost(x,y,w,b):
    m = len(y)
    cost = 0
    for i in range(0,m):
        z_i =  np.dot(x[i],W) + b
        #print(type(z_i))
        f_wb_i = sigmoid(z_i)
        cost = cost + (-y[i]*np.log(f_wb_i) - (1-y[i])*np.log(1 - f_wb_i))
        return cost/m
```

```python
print(compute_cost(x_train,y_train,W,1))
#x_train.shape
```

```python
def compute_decent(x,y,w,b):
    m,n = x.shape
    dj_dw = np.zeros((n,))
    dj_db = 0
    for i in range(0,m):
        f_wb = sigmoid(np.dot(x[i],w) +b)
        err = f_wb - y[i]
        for j in range(0,n):
            dj_dw[j] = dj_dw[j] + err*x[i][j]
        dj_db = dj_db + err
    dj_dw = dj_dw/m                                   
    dj_db = dj_db/m 
    return dj_db, dj_dw 
```
